### Testing

```python
age = 55
gender = 0
height_cm = 156
weight_kg = 85
ap_hi = 140
ap_lo = 90
cholesterol = 3
gluc = 1
smoke = 0
alco = 0
active = 1
cholesterol_gluc_interaction = gluc * cholesterol

height_m = height_cm / 100
bmi = weight_kg / (height_m ** 2)
pulse_pressure = ap_hi - ap_lo

chol_2 = 1 if cholesterol == 2 else 0
chol_3 = 1 if cholesterol == 3 else 0

gluc_2 = 1 if gluc == 2 else 0
gluc_3 = 1 if gluc == 3 else 0

input_data = pd.DataFrame([{
    'age_years': age,
    'gender': gender,
    'BMI': bmi,
    'ap_hi': ap_hi,
    'ap_lo': ap_lo,
    'pulse_pressure': pulse_pressure,
    'cholesterol_gluc_interaction': cholesterol_gluc_interaction,
    'smoke': smoke,
    'alco': alco,
    'active': active,
    'chol_2': chol_2,
    'chol_3': chol_3,
    'gluc_2': gluc_2,
    'gluc_3': gluc_3
}])
```

```python
prediction = model.predict(input_data)[0]
probability = model.predict_proba(input_data)[0]
```

```python
if prediction == 1:
    label_text = "HIGH RISK"
    probability_text = round(probability[1] * 100, 2)
    display_text = f"{label_text} ({probability_text}% chance of disease)"
else:
    label_text = "LOW RISK"
    probability_text = round((1 - probability[1]) * 100, 2)
    display_text = f"{label_text} ({probability_text}% chance of no disease)"
```
